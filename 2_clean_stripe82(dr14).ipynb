```python
import pyarrow as pa
import pyarrow.parquet as pq
import pandas as pd
```

```python
path = '/content/drive/MyDrive/DL/Data/s82ObjectTable.parquet'

obj = pd.read_parquet(path, engine = 'pyarrow')
```

```python
obj
#살릴 항목: objectId, ra, dec, psMag_ugriz
#추가할 항목: ug,gr,ri,iz,zu
```

```python
newobj = obj[['ra','dec','psMag_u', 'psMag_g','psMag_r', 'psMag_i','psMag_z']]
```

```python
newobj
```

```python
newobj['ug'] = newobj['psMag_u'] - newobj['psMag_g']
```

```python
newobj['gr'] = newobj['psMag_g'] - newobj['psMag_r']
newobj['ri'] = newobj['psMag_r'] - newobj['psMag_i']
newobj['iz'] = newobj['psMag_i'] - newobj['psMag_z']
newobj['zu'] = newobj['psMag_z'] - newobj['psMag_u']
```

```python
newobj = newobj.rename(columns={'u':'u_band'})
newobj = newobj.rename(columns={'g':'g_band'})
newobj = newobj.rename(columns={'r':'r_band'})
newobj = newobj.rename(columns={'i':'i_band'})
newobj = newobj.rename(columns={'z':'z_band'})
```

```python
newobj
```

```python
newobj.to_csv('/content/drive/MyDrive/DL/Data/clean_LC.csv')
```
