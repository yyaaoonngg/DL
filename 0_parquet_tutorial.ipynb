```python
import pyarrow as pa
import pyarrow.parquet as pq
import pandas as pd
```

```python
path1 = '/content/drive/MyDrive/DL/Data/s82MultiWavelengthTable.parquet'
path2 = '/content/drive/MyDrive/DL/Data/s82ObjectTable.parquet'
path3 = '/content/drive/MyDrive/DL/Data/s82SourceTable.parquet'

s82_Wl = pd.read_parquet(path1, engine = 'pyarrow')
S82_Obj = pd.read_parquet(path2, engine = 'pyarrow')
s82_Sc = pd.read_parquet(path3, engine = 'pyarrow')
```

```python
def get_parquet_columns(parquet_file_path):
    parquet_file = pq.ParquetFile(parquet_file_path)
    column_names = parquet_file.schema.names
    return column_names

columns1 = get_parquet_columns(path1)
columns2 = get_parquet_columns(path2)
columns3 = get_parquet_columns(path3)
```
